```python
%pylab inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = [10,100,1000,10000,100000,1000000]
tree_av = np.array(x)

tree_av = [440.132,
770.862,
1540.64,
1722.91,
2613.8,
2657.82,
]
tree_bal_av = [381.29,
567.288,
997.139,
935.022,
1706.48,
1996.81,
]

map_av = [152.63,
194.697,
278.494,
380.869,
443.08,
576.636,

]          
tree_av = np.array(tree_av)
tree_bal_av = np.array(tree_bal_av)
map_av = np.array(map_av)
fit = numpy.polyfit(numpy.log(x)/log(2), tree_bal_av, 1)
```

```python
semilogx(x,tree_av)
semilogx(x,tree_bal_av)
semilogx(x,map_av)
semilogx(x,fit[0]*log(x)/log(2)+fit[1])
ylabel('Time (ms) ')
xlabel('Size (N)')
title('Lookup time')

legend(['Tree unbalanced', 'Tree balanced', 'Map', 'Log2(x)'])
savefig('lookup_time.pdf')
```
